```python
# Anyone who has the same plots and data as me would be immediately reported. Please don't copy
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Parameters for normal distribution
mean = 200
std_dev = 30
n_values = [5, 10, 20, 40, 100, 1000]

# Prepare to store means and standard deviations for plotting
means = []
std_devs = []
n_list = []

# Confidence intervals (95% and 99%)
conf_95 = []
conf_99 = []

# For each n, generate normally distributed data and compute mean, std dev, and confidence intervals
for n in n_values:
    data = np.random.normal(mean, std_dev, n)
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)
    
    # Confidence intervals
    ci_95 = stats.t.interval(0.95, df=n-1, loc=sample_mean, scale=sample_std/np.sqrt(n))
    ci_99 = stats.t.interval(0.99, df=n-1, loc=sample_mean, scale=sample_std/np.sqrt(n))
    
    means.append(sample_mean)
    std_devs.append(sample_std)
    conf_95.append(ci_95)
    conf_99.append(ci_99)
    n_list.append(n)

    # Plot histogram for the current n
    plt.figure(figsize=(6, 4))
    plt.hist(data, bins=10, alpha=0.75, color='blue', edgecolor='black')
    plt.title(f'Histogram for n = {n}')
    plt.xlabel('Data Values')
    plt.ylabel('Frequency')
    plt.show()

# Plot mean and standard deviation as a function of n with error bars
plt.figure(figsize=(10, 6))

# Convert lists to numpy arrays for easier indexing
means = np.array(means)
std_devs = np.array(std_devs)
conf_95 = np.array(conf_95)
conf_99 = np.array(conf_99)

# Plot mean with error bars (95% confidence intervals)
plt.errorbar(n_list, means, yerr=[means - conf_95[:, 0], conf_95[:, 1] - means], fmt='o-', label='Mean with 95% CI')

# Plot mean with error bars (99% confidence intervals)
plt.errorbar(n_list, means, yerr=[means - conf_99[:, 0], conf_99[:, 1] - means], fmt='s-', label='Mean with 99% CI', color='orange')

# Plot standard deviation as a function of n
plt.plot(n_list, std_devs, 'x--', label='Standard Deviation')

plt.xscale('log')
plt.xlabel('n (log scale)')
plt.ylabel('Value')
plt.title('Mean and Standard Deviation as a Function of n with 95% and 99% CI')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd

# Create a DataFrame to report the results for each value of n
report_data = {
    'n': n_list,
    'Mean': means,
    'Standard Deviation': std_devs,
    '95% CI Lower': conf_95[:, 0],
    '95% CI Upper': conf_95[:, 1],
    '99% CI Lower': conf_99[:, 0],
    '99% CI Upper': conf_99[:, 1]
}

report_df = pd.DataFrame(report_data)

# Display the DataFrame
print(report_df)
```
